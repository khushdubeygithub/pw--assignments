When creating a custom exception in Python, it is common practice to inherit from the Exception class or one of its subclasses. There are a few reasons why it is recommended to use the Exception class as the base class for custom exceptions:
Consistency
Compatibility
Exception Hierarchy
Functionality

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError

```python
my_dict = {"apple": "red", "banana": "yellow"}

try:
    value = my_dict["grape"]
except KeyError as e:
    print("Error:", e)
```

```python
my_list = [1, 2, 3]

try:
    value = my_list[3]
except IndexError as e:
    print("Error:", e)
```

Q5. Explain ImportError. What is ModuleNotFoundError

ImportError is an exception that is raised when an import statement fails to locate and import a module or when there is an issue with importing a module. It is a subclass of the built-in Exception class

```python
try:
    import nonexistent_module
except ImportError as e:
    print("ImportError occurred:", e)

try:
    import nonexistent_module2
except ModuleNotFoundError as e:
    print("ModuleNotFoundError occurred:", e)
```
