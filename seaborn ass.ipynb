Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Scatter Plot: A scatter plot is used to visualize the relationship between two continuous variables. It helps in identifying patterns, clusters, or outliers in data.

Line Plot: A line plot is typically used to display data points in a time series or to show the trend or change in data over a continuous interval.

Bar Plot: A bar plot is used to compare the values of different categories or groups. It is effective for displaying discrete data and making comparisons.

Box Plot: A box plot, also known as a box-and-whisker plot, is used to display the distribution of a dataset. It shows summary statistics like median, quartiles, and potential outliers.

Histogram: A histogram is used to visualize the distribution of a single variable by dividing it into bins and counting the number of observations in each bin. It provides insights into the data's shape and central tendency.

Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Create box plots
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x="pclass", y="age", data=titanic_data)
plt.title("Box Plot of Age by Passenger Class")

plt.subplot(1, 2, 2)
sns.boxplot(x="pclass", y="fare", data=titanic_data)
plt.title("Box Plot of Fare by Passenger Class")

plt.tight_layout()
plt.show()
```

Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram with hue
sns.histplot(data=diamonds_data, x="price", hue="cut", bins=30, kde=True)
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Histogram of Diamond Prices by Cut")
plt.show()
```

Q5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix for the heatmap
pivot_data = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
sns.heatmap(pivot_data, cmap="YlGnBu", annot=True, fmt="d")
plt.title("Heatmap of Flight Passengers")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
